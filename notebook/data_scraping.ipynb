```python
import os
import requests
from bs4 import BeautifulSoup
```

```python
BASE_URL = "https://cag.gov.in"
STATE_ID = 93
STATE_NAME = "Telangana"

BASE_SAVE_DIR = f"../data/raw/{STATE_NAME}"

url = f"{BASE_URL}/en/state-accounts-report?defuat_state_id={STATE_ID}"

response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

tab = soup.find("div", id="tab-360")

year_blocks = tab.find_all("div", class_="accTrigger")

download_count = 0

for year_block in year_blocks:
    year = year_block.get_text(strip=True).replace(" ", "")
    year_dir = os.path.join(BASE_SAVE_DIR, year)

    os.makedirs(year_dir, exist_ok=True)

    month_container = year_block.find_next_sibling("div", class_="accordDetail")

    for li in month_container.find_all("li"):
        month_name = li.find("h5").get_text(strip=True).replace(", ", "_")
        pdf_url = BASE_URL + li.find("a", href=True)["href"]

        file_name = f"{month_name}.pdf"
        file_path = os.path.join(year_dir, file_name)

        if os.path.exists(file_path):
            continue  # Skip if already downloaded

        print(f"Downloading: {year}/{file_name}")

        pdf_data = requests.get(pdf_url).content
        with open(file_path, "wb") as f:
            f.write(pdf_data)

        download_count += 1

print(f"\nTotal PDFs downloaded: {download_count}")
```
